```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
cateogical_feature_name = ['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']
for feature_name in cateogical_feature_name:
    data[feature_name] = label_encoder.fit_transform(data[feature_name])
```

```python
#multiple linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
def multiple_reression(target_feature, other_feature, data, x_train, x_test, y_train, y_test):
    
    multi_linear_model = LinearRegression()
    multi_linear_model.fit(x_train,y_train)
    y_predict = multi_linear_model.predict(x_test)
    score = multi_linear_model.score(x_train, y_train)
    plt.title('multiple regression graph')
    plt.scatter(y_predict,y_test)
    plt.xlabel('predict')
    plt.ylabel('real value')
    plt.show()
    
    temp = pd.DataFrame()
    temp['y_predict'] = y_predict
    list_temp = list(y_test)
    temp['y_test'] = list_temp
    index = [i for i in range(0, len(temp))]
    temp['index'] = index
    plt.figure(figsize=(20, 5))
    plt.title('multiple regression graph2')
    plt.scatter(temp['index'],temp['y_predict'], color = 'red', alpha = 0.3, label = 'predict')
    plt.scatter(temp['index'],temp['y_test'], color = 'blue', alpha = 0.3, label = 'real value')
    plt.legend()
    plt.show()
    
    return score
x = data[other_feature]
y = data[target_feature]
    
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2)
mp_score = multiple_reression(target_feature, other_feature, data, x_train, x_test, y_train, y_test)
print('score of multiple regression :', mp_score)
```

```python
#polynomial regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
def polynomial_regression(target_feature, other_feature, user_degree, degree, data, x_train, x_test, y_train, y_test):

    poly = PolynomialFeatures(degree = user_degree, include_bias = True)
    x_train_poly = poly.fit_transform(x_train)

    line_reg = LinearRegression()
    line_reg.fit(x_train_poly, y_train)

    x_test_poly = poly.transform(x_test)
    y_predict = line_reg.predict(x_test_poly)
    score = line_reg.score(x_train_poly, y_train)
    
    plt.title('polynomial regression graph')
    plt.scatter(y_predict,y_test, label = "real price")
    plt.xlabel('predict')
    plt.ylabel('real value')
    plt.show()
    
    temp = pd.DataFrame()
    temp['y_predict'] = y_predict
    list_temp = list(y_test)
    temp['y_test'] = list_temp
    index = [i for i in range(0, len(temp))]
    temp['index'] = index   
    plt.figure(figsize=(20, 5))
    plt.title('polynomial regression graph2')
    plt.scatter(temp['index'],temp['y_predict'], color = 'red', alpha = 0.3, label = 'predict')
    plt.scatter(temp['index'],temp['y_test'], color = 'blue', alpha = 0.3, label = 'real value')
    plt.legend()
    plt.show()
    
    return score;
x = data[other_feature]
y = data[target_feature]
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2)
score = polynomial_regression(target_feature, other_feature, 2, data, x_train, x_test, y_train, y_test)
print('score of polynomial regression :', score)
```
